```python
import sympy as sp



```

```python
z=sp.symbols('z',positive=True,real=True)
C=sp.symbols('C_F',constant=True)
P=C*((1+(1-z)**2)/z)
pt,E=sp.symbols('p_T,E',positive=True,real=True)
zp=1-pt/E
zm=pt/E
I=sp.integrate(P,(z,zp,zm)).simplify().doit()
pmax,pmin=sp.symbols('ptmax,ptmin',positive=True,real=True)
I
```

```python
F=sp.Function('F')(pt)
F=(2/pt)*I
F
```

```python
sp.integrate(F,(pt,pmin,pmax)).doit()
```
